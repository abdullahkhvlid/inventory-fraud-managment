```python

inventory = {
    "apple": 30,
    "banana": 20,
    "peach": 40,
    "pineapple": 20,
    "kiwi": 43,
    "mango": 86
}

cart = {}
returns = {}
purchase_history = []
return_history = []

def show_inventory():
    print("\nAvailable Inventory:")
    for item, qty in inventory.items():
        print(f" - {item}: {qty}")

def purchase_item():
    while True:
        item = input("\nWhat do you want to buy? (type 0 to stop, 1 to return items): ").strip().lower()

        if item == "0":
            break
        elif item == "1":
            return_items()
            continue

        if item not in inventory:
            print("This item is not available in our store.")
            continue

        try:
            quantity = int(input("Quantity? "))
            if quantity <= 0:
                print("Quantity must be greater than 0.")
                continue
            if quantity > inventory[item]:
                print(f"Only {inventory[item]} items available.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        cart[item] = cart.get(item, 0) + quantity
        inventory[item] -= quantity
        purchase_history.append(item)
        print(f"{quantity} {item}(s) added to your cart.")

        if len(purchase_history) > 5:
            print("Note: You're purchasing more than 5 items.")

def return_items():
    print("\n-- Return Mode Activated --")
    while True:
        item = input("Which item do you want to return? (type 3 to stop): ").strip().lower()
        if item == "3":
            break

        if item not in inventory:
            print("This item is not part of our inventory.")
            continue

        try:
            quantity = int(input("Quantity to return? "))
            if quantity <= 0:
                print("Quantity must be greater than 0.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        returns[item] = returns.get(item, 0) + quantity
        inventory[item] += quantity
        return_history.append(item)
        print(f"{quantity} {item}(s) returned successfully.")

        if len(return_history) > 5:
            print("Warning: This customer is returning more than 5 items.")

def show_summary():
    print("\n--- Transaction Summary ---")
    print("Cart Items:")
    for item, qty in cart.items():
        print(f" - {item}: {qty}")

    print("\nReturned Items:")
    for item, qty in returns.items():
        print(f" - {item}: {qty}")

    print("\nRemaining Inventory:")
    for item, qty in inventory.items():
        print(f" - {item}: {qty}")


if __name__ == "__main__":
    print("Welcome to the Inventory System")
    show_inventory()
    purchase_item()
    show_summary()
```
